```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

```python
# Load the data
df = pd.read_csv("assets/epa-sea-level.csv")
df
```

Use matplotlib to create a scatter plot using the Year column as the x-axis and the CSIRO Adjusted Sea Level column as the y-axis.

```python
# Create scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df['Year'], df['CSIRO Adjusted Sea Level'], color='b', label='Sea Level')

# Add labels and title
plt.xlabel('Year')
plt.ylabel('CSIRO Adjusted Sea Level (inches)')
plt.title('Sea Level Rise')

# Show plot
plt.legend()
plt.show()
```

Use the linregress function from scipy.stats to get the slope and y-intercept of the line of best fit. Plot the line of best fit over the top of the scatter plot. Make the line go through the year 2050 to predict the sea level rise in 2050.

```python
# Create scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df['Year'], df['CSIRO Adjusted Sea Level'], color='b', label='Sea Level')

# Perform linear regression
slope, intercept, r_value, p_value, std_err = linregress(df['Year'], df['CSIRO Adjusted Sea Level'])

# Create an array of years from the start year to 2050
years = np.arange(df['Year'].min(), 2051)

# Calculate the corresponding sea levels using the slope and intercept
sea_levels = intercept + slope * years

# Plot the line of best fit
plt.plot(years, sea_levels, color='r', label='Best Fit Line')

# Add labels and title
plt.xlabel('Year')
plt.ylabel('CSIRO Adjusted Sea Level (inches)')
plt.title('Sea Level Rise')
plt.legend()

# Show plot
plt.show()
```
